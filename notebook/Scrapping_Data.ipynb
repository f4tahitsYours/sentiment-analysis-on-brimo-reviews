```python
from google_play_scraper import reviews, Sort
import pandas as pd
```

```python
def scrape_brimo_reviews(count=20000, save_to_csv=True, filename='ulasan_brimo.csv'):
    try:
        print("üîç Mengambil ulasan BRImo dari Google Play Store...")

        # Scraping ulasan
        result, _ = reviews(
            'id.co.bri.brimo',
            lang='id',
            country='id',
            sort=Sort.NEWEST,
            count=count,
            filter_score_with=None
        )

        # Konversi hasil ke DataFrame
        df = pd.DataFrame(result)

        # Simpan ke CSV jika diinginkan
        if save_to_csv:
            df.to_csv(filename, index=False)
            print(f"‚úÖ {len(df)} ulasan berhasil disimpan ke '{filename}'")

        return df

    except Exception as e:
        print(f"‚ùå Terjadi kesalahan saat mengambil ulasan: {e}")
        return None
```

```python
# Panggil fungsi scraping dan simpan hasil ke variabel df
df = scrape_brimo_reviews()

if df is not None:
    df.head()
```
